## Main Imports

```python
#main imports
from numpy import*
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
```

## Constant Variables 

```python
#variables for conversion from seconds to min and hours
stm=60
sth=3600
```

## 20ËšC Test

```python
#20 degreed Celcius trial

#use the Pandas library to read the data
datafile_20 = pd.read_csv('04-14 50 PSI 20 SLPM 20 C .csv')

#create arrays with the Mass Flow data
MF_20=array([datafile_20['MASSFLOW']])
#convert the Mass Flow from SLPM to SLPS ()
MF_20=MF_20/stm

#flaten MF_20 so it has the same dimension that linspace will create
MF_20=MF_20.flatten()

#set x values
L=MF_20.size
X=linspace(0,L,L)

#integrate under the curve to obtain the Total Mass
Int_20=sp.integrate.cumulative_trapezoid(MF_20,X)
#isolate the total mass
TM_20=Int_20[-1]

#unsure if this will stay
#create a plot of the Mass Flow over Time
plt.plot(X,MF_20,'gold')
plt.ylabel('Mass Flow (SLPS)')
plt.xlabel('Time (Seconds)')
plt.title('Mass Flow vs. Time 20ËšC')
plt.show()

#create a plot of the Total Mass over Time (the integral is one less figure than the input so need to adjust for graph
plt.plot(X[1:],Int_20,'orange')
plt.ylabel('Total Mass (Liters)')
plt.xlabel('Time (Seconds)')
plt.title('Total Mass vs. Time at 20ËšC')
plt.show()

print('The total mass is approximately {:.2f} L at 20ËšC'.format(TM_20))

#Math to calculate how long it takes to reach equalibrium approx. 90% of total mass
eq=.9*TM_20
#find the position of eq in TM
above_eq=where(Int_20>=eq) 
eq_pos=above_eq[0][0]  
#convert from seconds
eq_T_20=eq_pos/stm
eq_T_20_h=eq_pos/sth
print('Equalibrium occures after approximately {:.2f} min or {:.2f} hours at 20ËšC'.format(eq_T_20,eq_T_20_h))

#calculating an approximate error on the total mass (this is due to the integration method)
#this portion could be expanded to average multiple tests at a single temperature for increased accuracy
err_20=TM_20*.1
err_20_T=eq_T_20_h*.1
```

## Intergral Method Comparison

```python
from scipy.integrate import simpson
from numpy import trapz

#comparing the result of several methods of integration
#there is no significant improvement in accuracy of integration using diffrent methods

#use the Pandas library to read the data
datafile = pd.read_csv('0501 50PSI 20 SLPM 5C .csv')

#create arrays with the Mass Flow data
MF=array([datafile['MASSFLOW']])
#convert the Mass Flow from SLPM to SLPS ()
MF=MF/stm
#flaten MF so it has the same dimension that linspace will create
MF=MF.flatten()

L=MF.size
X=linspace(0,L,L)

# Compute the area using the cumulative trapezoidal rule.
area=sp.integrate.cumulative_trapezoid(MF,X)
print("area using cumulative trap, stepsize X =", area[-1])

# Compute the area using the composite trapezoidal rule.
area = trapz(MF, dx=1)
print("\narea using composite trap, stepsize 1 =", area)

# Compute the area using the composite trapezoidal rule.
area = trapz(MF, dx=.9)
print("area using composite trap, stepsize .9 =", area)

# Compute the area using the composite trapezoidal rule.
area = trapz(MF, dx=1.1)
print("area using composite trap, stepsize 1.1 =", area)

# Compute the area using the composite Simpson's rule.
area = simpson(MF, X)
print("\narea using simpsons, stepsize X =", area)

# Compute the area using the composite Simpson's rule.
area = simpson(MF, dx=1)
print("area using simpsons, stepsize 1 =", area)

# Compute the area using the composite Simpson's rule.
area = simpson(MF, dx=.9)
print("area using simpsons, stepsize .9 =", area)

# Compute the area using the composite Simpson's rule.
area = simpson(MF, dx=1.1)
print("area using simpsons, stepsize 1.1 =", area)
```
