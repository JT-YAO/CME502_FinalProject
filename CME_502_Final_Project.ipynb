```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

df = pd.read_csv('B:\SBU\OneDrive - Stony Brook University\CTSR\Thesis\ICP Data\CSV Files for Program\ICPs_for_Thesis.csv')

Powder = df['Powder']
Chemistry = df['Chemistry']
Matrix_Mass = df['Matrix Mass %'].to_numpy()
Matrix_Vol = df['Matrix Volume %'].to_numpy()
Stoich = df['Stoichiometry'].to_numpy()
Velocity = df['Velocity'].to_numpy()
Temp = df['Temperature'].to_numpy()
Momentum = df['Momentum'].to_numpy()
Kinetic = df['Kinetic Energy'].to_numpy()
Evolving = df['Evolving Stress'].to_numpy()
Residual = df['Residual Stress'].to_numpy()
Radius = df['Particle Radius'].to_numpy()
Latent = df['Matrix Latent Heat of Fusion'].to_numpy()
Matrix_Group = df['Matrix Group']
Momentum_Matrix = Momentum*Matrix_Vol
Momentum_Matrix_Latent = Momentum_Matrix*Latent
Kinetic_Energy_Matrix = Kinetic*Matrix_Vol

def linear(x,a,b):
    return a*x+b
    
markers = ['o','o','o','^','^','^','s','s','s','x','x','x','D','D','D','P','P','P','v','v','v']
colors = ['b','b','b','g','g','g','r','r','r','m','m','m','y','y','y','k','k','k','tab:orange','tab:orange','tab:orange',]
units = [None]*24
unit_append = ['kg-m/s','J','kg-m/s','J','kg']
units = units + unit_append
units[14] = 'm/s'

parameters = [14,24,25,26,27,28]
for k in parameters:
    i = 0
    while i < 9:
        plt.scatter(df[df.columns[k]][i:i+3],Evolving[i:i+3],\
                    label=Chemistry[i],marker=markers[i],c=colors[i])
        i = i + 3
    t = np.linspace(min(df[df.columns[k]][0:9]),\
                    max(df[df.columns[k]][0:9]))
    constants, pcov = curve_fit(linear, df[df.columns[k]][0:9], Evolving[0:9])
    residuals = Evolving[0:9] - linear(df[df.columns[k]][0:9], *constants)
    ss_res = np.sum(residuals**2)
    ss_tot = np.sum((Evolving[0:9]-np.mean(Evolving[0:9]))**2)
    r_squared = 1 - (ss_res / ss_tot)
    a_fit = constants[0]
    b_fit = constants[1]
    fit = []
    for j in t:
        fit.append(linear(j,a_fit,b_fit))
    plt.plot(t,fit)
    plt.annotate('y = {:.2e}x + {:.2f}'.format(a_fit,b_fit),(70,65),xycoords = 'figure points')
    plt.annotate('R^2 = {:.2f}'.format(r_squared),(70,50),xycoords = 'figure points')
    plt.title(Matrix_Group[i-1] + ' ' + df.columns[k] + '\n vs. Evolving Stress')
    plt.xlabel(df.columns[k] + ' ' + units[k])
    plt.ylabel('Evolving Stress (MPa)')
    plt.legend()
    plt.ticklabel_format(axis='x', style='sci',scilimits=(0,0))
    plt.savefig(Matrix_Group[i-1] + '_' + df.columns[k] + '.jpg',bbox_inches = 'tight')
    plt.show()
    while i < 15:
        plt.scatter(df[df.columns[k]][i:i+3],Evolving[i:i+3],\
                    label=Chemistry[i],marker=markers[i],c=colors[i])
        i = i + 3
    t = np.linspace(min(df[df.columns[k]][9:15]),max(df[df.columns[k]][9:15]))
    constants, pcov = curve_fit(linear, df[df.columns[k]][9:15], Evolving[9:15])
    residuals = Evolving[9:15] - linear(df[df.columns[k]][9:15], *constants)
    ss_res = np.sum(residuals**2)
    ss_tot = np.sum((Evolving[9:15]-np.mean(Evolving[9:15]))**2)
    r_squared = 1 - (ss_res / ss_tot)
    a_fit = constants[0]
    b_fit = constants[1]
    fit = []
    for j in t:
        fit.append(linear(j,a_fit,b_fit))
    plt.plot(t,fit)
    plt.annotate('y = {:.2e}x + {:.2f}'.format(a_fit,b_fit),(70,65),xycoords = 'figure points')
    plt.annotate('R^2 = {:.2f}'.format(r_squared),(70,50),xycoords = 'figure points')
    plt.title(Matrix_Group[i-1] + ' ' + df.columns[k] + '\n vs. Evolving Stress')
    plt.xlabel(df.columns[k] + ' ' + units[k])
    plt.ylabel('Evolving Stress (MPa)')
    plt.legend()
    plt.ticklabel_format(axis='x', style='sci',scilimits=(0,0))
    plt.savefig(Matrix_Group[i-1] + '_' + df.columns[k] + '.jpg',bbox_inches = 'tight')
    plt.show()
    while i < 18:
        plt.scatter(df[df.columns[k]][i:i+3],Evolving[i:i+3],\
                    label=Chemistry[i],marker=markers[i],c=colors[i])
        i = i + 3
    t = np.linspace(min(df[df.columns[k]][15:18]),max(df[df.columns[k]][15:18]))
    constants, pcov = curve_fit(linear, df[df.columns[k]][15:18], Evolving[15:18])
    residuals = Evolving[15:18] - linear(df[df.columns[k]][15:18], *constants)
    ss_res = np.sum(residuals**2)
    ss_tot = np.sum((Evolving[15:18]-np.mean(Evolving[15:18]))**2)
    r_squared = 1 - (ss_res / ss_tot)
    a_fit = constants[0]
    b_fit = constants[1]
    fit = []
    for j in t:
        fit.append(linear(j,a_fit,b_fit))
    plt.plot(t,fit)
    plt.annotate('y = {:.2e}x + {:.2f}'.format(a_fit,b_fit),(70,65),xycoords = 'figure points')
    plt.annotate('R^2 = {:.2f}'.format(r_squared),(70,50),xycoords = 'figure points')
    plt.title(Matrix_Group[i-1] + ' ' + df.columns[k] + '\n vs. Evolving Stress')
    plt.xlabel(df.columns[k] + ' ' + units[k])
    plt.ylabel('Evolving Stress (MPa)')
    plt.legend()
    plt.ticklabel_format(axis='x', style='sci',scilimits=(0,0))
    plt.savefig(Matrix_Group[i-1] + '_' + df.columns[k] + '.jpg',bbox_inches = 'tight')
    plt.show()
    while i < 21:
        plt.scatter(df[df.columns[k]][i:i+3],Evolving[i:i+3],\
                    label=Chemistry[i],marker=markers[i],c=colors[i])
        i = i + 3
    t = np.linspace(min(df[df.columns[k]][18:21]),max(df[df.columns[k]][18:21]))
    constants, pcov = curve_fit(linear, df[df.columns[k]][18:21], Evolving[18:21])
    residuals = Evolving[18:21] - linear(df[df.columns[k]][18:21], *constants)
    ss_res = np.sum(residuals**2)
    ss_tot = np.sum((Evolving[18:21]-np.mean(Evolving[18:21]))**2)
    r_squared = 1 - (ss_res / ss_tot)
    a_fit = constants[0]
    b_fit = constants[1]
    fit = []
    for j in t:
        fit.append(linear(j,a_fit,b_fit))
    plt.plot(t,fit)
    plt.annotate('y = {:.2e}x + {:.2f}'.format(a_fit,b_fit),(70,65),xycoords = 'figure points')
    plt.annotate('R^2 = {:.2f}'.format(r_squared),(70,50),xycoords = 'figure points')
    plt.title(Matrix_Group[i-1] + ' ' + df.columns[k] + '\n vs. Evolving Stress')
    plt.xlabel(df.columns[k] + ' ' + units[k])
    plt.ylabel('Evolving Stress (MPa)')
    plt.legend()
    plt.ticklabel_format(axis='x', style='sci',scilimits=(0,0))
    plt.savefig(Matrix_Group[i-1] + '_' + df.columns[k] + '.jpg',bbox_inches = 'tight')
    plt.show()
```
