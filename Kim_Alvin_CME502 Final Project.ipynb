```python
import pandas as pd
from tkinter.filedialog import askopenfilenames as ask
import matplotlib.pyplot as plt

cp = ask() # User can select the desired sample files

max_load = [] 
Adhesion_list =[]   
for i in cp:
    df = pd.read_csv(i) # reads all the selected csv files
    
    pos = df["Position(8800 (0,1):Position) (mm)"]
    load = df["Load(8800 (0,1):Load) (kN)"]
    
    maxload = load.max()
    max_load.append(maxload)
   
    N = maxload*1000
    Adhesion = N/490.87
    Adhesion_list.append(Adhesion)
    average = sum(Adhesion_list)/len(Adhesion_list)
    
    plt.plot(pos,load)
    plt.title(i)
    plt.xlabel('Position (mm)')
    plt.ylabel('Load (kN)')
    plt.show()

for i in range (0, len(max_load)):
     print("The max load for file {}: {:.2f} N\n".format(i,max_load[i]))
for i in range (0, len(Adhesion_list)):
     print("The adhesion strength for file {}: {:.2f} MPa\n".format(i,Adhesion_list[i]))
print("The average adhesion strength is {:.2f} MPa".format(average))
```
